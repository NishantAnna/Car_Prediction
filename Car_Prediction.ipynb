```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.cluster import KMeans
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings("ignore")

# --- Load Data ---
df = pd.read_csv("Final_Used_Cars.csv")

# --- Feature Engineering ---

# 1. Car age
current_year = datetime.now().year
df['car_age'] = current_year - df['year']

# 2. Price per mile (avoid division by zero)
df['price_per_mile'] = df['price'] / df['miles'].replace(0, 1)

# 3. Region clustering using KMeans on latitude and longitude
coords = df[['latitude', 'longitude']].to_numpy()
kmeans = KMeans(n_clusters=10, random_state=42)
df['region_cluster'] = kmeans.fit_predict(coords).astype(str)  # as categorical for encoding

# --- Prepare Features and Target ---
X = df.drop(columns=["price"])
y = df["price"]

# --- Identify categorical and numeric columns ---
cat_cols = X.select_dtypes(include=['object']).columns.tolist()
num_cols = X.select_dtypes(exclude=['object']).columns.tolist()

# --- Preprocessing Pipeline ---
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
])

# --- Define XGBoost Regressor and Pipeline ---
xgb_model = XGBRegressor(objective='reg:squarederror', random_state=42, verbosity=0)

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', xgb_model)
])

# --- Train-Test Split ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Hyperparameter distributions for Randomized Search ---
param_dist = {
    'regressor__n_estimators': [100, 200, 300, 400],
    'regressor__max_depth': [4, 6, 8, 10, 12],
    'regressor__learning_rate': [0.01, 0.05, 0.1, 0.15],
    'regressor__subsample': [0.6, 0.8, 1.0],
    'regressor__colsample_bytree': [0.6, 0.8, 1.0],
    'regressor__gamma': [0, 0.1, 0.3, 0.5],
    'regressor__reg_lambda': [1, 1.5, 2, 3],
    'regressor__min_child_weight': [1, 3, 5],
}

# --- Randomized Search with 5-fold Cross Validation ---
random_search = RandomizedSearchCV(
    estimator=model_pipeline,
    param_distributions=param_dist,
    n_iter=50,
    scoring='neg_mean_absolute_error',
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# --- Fit ---
random_search.fit(X_train, y_train)

# --- Best Model ---
best_model = random_search.best_estimator_
print("Best hyperparameters:")
print(random_search.best_params_)

# --- Predict and Evaluate ---
y_pred = best_model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE) on Test Set: {mae:.2f}")
print(f"R² Score on Test Set: {r2:.4f}")

# --- Sample Predictions ---
print("\nSample predictions (first 10):")
for actual, pred in zip(y_test.head(10), y_pred[:10]):
    print(f"Actual: {actual:.0f} \tPredicted: {pred:.0f} \tDifference: {actual - pred:.0f}")
```

```python
# train_and_save.py
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, r2_score
from xgboost import XGBRegressor
import joblib
import warnings
warnings.filterwarnings("ignore")

# --- Load data ---
df = pd.read_csv("Final_Used_Cars.csv")
df = df.dropna(subset=["price"])  # make sure target exists

# --- Feature engineering (NO leakage) ---
current_year = datetime.now().year
df['car_age'] = current_year - df['year']

# Create region clusters (KMeans) from lat/lon and save kmeans for inference
coords = df[['latitude', 'longitude']].to_numpy()
kmeans = KMeans(n_clusters=10, random_state=42)
df['region_cluster'] = kmeans.fit_predict(coords).astype(str)

# NOTE: Do NOT create price_per_mile from price (target). If you want a "price_per_mile" proxy,
# compute a dataset-level median and use it as a fallback for production (we save median below)
median_ppm = (df['price'] / df['miles'].replace(0, np.nan)).median()
# We'll save median_ppm as reference (not used as a model feature directly here)
joblib.dump(median_ppm, "median_price_per_mile.pkl")

# --- Prepare X/y ---
X = df.drop(columns=["price"])
y = df["price"]

# Identify categorical and numeric columns
cat_cols = X.select_dtypes(include=['object']).columns.tolist()
num_cols = X.select_dtypes(exclude=['object']).columns.tolist()

# Build preprocessor and pipeline
preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
])

xgb = XGBRegressor(objective='reg:squarederror', random_state=42, verbosity=0)

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', xgb)
])

# Hyperparameter grid
param_dist = {
    'regressor__n_estimators': [100, 200, 300, 400],
    'regressor__max_depth': [4, 6, 8, 10],
    'regressor__learning_rate': [0.01, 0.05, 0.1],
    'regressor__subsample': [0.6, 0.8, 1.0],
    'regressor__colsample_bytree': [0.6, 0.8, 1.0],
    'regressor__gamma': [0, 0.1, 0.3],
    'regressor__reg_lambda': [1, 1.5, 2],
    'regressor__min_child_weight': [1, 3, 5],
}

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=40,
    scoring='neg_mean_absolute_error',
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)

random_search.fit(X_train, y_train)

best_model = random_search.best_estimator_
print("Best params:", random_search.best_params_)

# Evaluate
y_pred = best_model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MAE: {mae:.2f}, R2: {r2:.4f}")

# Save artifacts
joblib.dump(best_model, "car_price_model.pkl")
joblib.dump(kmeans, "kmeans_region.pkl")
# Save MAE for use in UI (to give ± MAE interval)
joblib.dump(mae, "test_mae.pkl")

print("Saved: car_price_model.pkl, kmeans_region.pkl, median_price_per_mile.pkl, test_mae.pkl")
```
